```python
import pandas as pd
from arch_comp_moonlight.utils import merge_results
```

```python
# MERGED_FILE = 'Moonlight_TuRBO_ARCH2024.csv'
MERGED_FILE = None
df = merge_results('output/**/*.csv', to_file=MERGED_FILE)
df["R"] = df["simulation_time"] / df["total_time"] * 100
df
```

```python
def falsification_rate(series: pd.Series) -> float:
    return (series == "yes").sum()

df2 = df.groupby(['instance', 'system', 'property']).agg({ # type: ignore
    'simulations': ['mean', "median"],
    'R': 'mean',
    'falsified': falsification_rate
})

df2
```
